```python
#reading Dataset
imdb_data= pd.read_csv("/home/ubuntu/Desktop/Assignment/Movie-Review-Sentiment-Analysis/imdb_1000.csv")
```

```python
#With best public review or with critics rating
sorted_imdb_data = imdb_data.sort_values(by='star_rating', ascending=False)

# Display the top-rated movies (e.g., top 10)
top_rated_movies = sorted_imdb_data.head(5)
print(top_rated_movies)
```

```python
x=imdb_data.title.head(10)
y=imdb_data.duration.head(10)
plt.scatter(y,x)
plt.savefig('scatter_plot.png')
plt.show()
```

```python
genre_count=imdb_data['genre'].value_counts()
print(genre_count)
```

```python
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                        plt.figure(figsize=(10,6))
genre_count.plot(kind='bar',color='skyblue')
plt.title('count of movies per genre')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.show()
```

```python
x = imdb_data['title'].head(10)
y = imdb_data['duration'].head(10)

# Create a pie chart
plt.figure(figsize=(8, 8))  # Adjust the figure size as needed
plt.pie(y, labels=x, autopct='%1.1f%%', startangle=140)

# Add a title
plt.title('Pie Chart of Top 10 Movie Durations')

# Save the pie chart as an image
plt.savefig('pie_chart.png')

# Display the pie chart
```
